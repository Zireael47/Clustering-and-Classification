```python
import pandas as pd
df = pd.read_csv("texture.dat")
df.head()
```

Датасет содержит 40 признаков А1-А40 и метку класса в столбце Class

```python
df.info()
```

Все типы данных подобраны правильно - все числовые

```python
df.groupby('Class').Class.count()
```

Все классы одинакового размера

```python
df.describe()
```

Стандартное отклонение для признаков (std) также отличается не сильно

```python
# стандартизируем данные
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[df.columns[:-1]])
df_scaled
```

```python
# расчитаем инерцию (inertia), искажение (distortion) и метрику силуэта (silhouette)
from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist
from sklearn.metrics import silhouette_score
import numpy as np

inertia=[]
distortion = []
silhouette = []
rg = range(2, 1)
for k in rg:
    kmeans = KMeans(n_clusters = k, random_state = 22222).fit(df_scaled)
    inertia.append(kmeans.inertia_)
    distortion.append(sum(np.min(cdist(df_scaled,
        kmeans.cluster_centers_,
        'euclidean'),axis = 1))/df_scaled.shape[0])
    preds = kmeans.predict(df_scaled)
    silhouette.append(silhouette_score(df_scaled,
    preds))
print('inertia: ', inertia)
print('distortion: ', distortion)
print('silhouette: ', silhouette)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("Метод локтя для инерции\n",fontsize=16)
plt.scatter(x=[i for i in rg],y=inertia,s=150,edgecolor='k')
plt.plot(rg, inertia);
plt.grid(True)
plt.xlabel("Число кластеров",fontsize=14)
plt.ylabel("Инерция",fontsize=15)
plt.xticks([i for i in rg],fontsize=14)
plt.yticks(fontsize=15)
plt.show()
```

По методу локтя для инерции оптимальное число кластеров 10, так как только после этой точки график приближается к прямой.

```python
plt.title("Метод локтя для иcкажения\n",fontsize=16)
plt.scatter(x=[i for i in rg],
y=distortion,s=150,edgecolor='k')
plt.plot(rg, distortion);
plt.grid(True)
plt.xlabel("Число кластеров",fontsize=14)
plt.ylabel("Искажение",fontsize=15)
plt.xticks([i for i in rg],fontsize=14)
plt.yticks(fontsize=15)
plt.show()
```

По методу локтя для искажения оптимальное число кластеров равно примерно 9, так как после этой точки график приближается к прямой.

```python
plt.title("Силуэт\n",fontsize=16)
plt.scatter(x=[i for i in rg],
y=silhouette,s=150,edgecolor='k')
plt.plot(rg,silhouette);
plt.grid(True)
plt.xlabel("Число кластеров",fontsize=14)
plt.ylabel("Силуэт",fontsize=15)
plt.xticks([i for i in rg],fontsize=14)
plt.yticks(fontsize=15)
plt.show()
```

По графику силуэта оптимальное число кластеров равно 2. Вероятно, что кластеры перемешаны, поэтому метод силуэта считает, что их разделять не надо.

```python
print("Real number of clusters: ", len(df['Class'].unique()))
```

```python
n_clusters = len(df.iloc[:, -1].unique())
kmeans = KMeans(n_clusters = n_clusters, random_state = 22222).fit(df_scaled)
df['kmeans_labels'] = kmeans.labels_
table = pd.crosstab(df['Class'], df['kmeans_labels'])
table
```

Из кросс-таблицы видно, что k-means некоторые кластеры разбил на несколько (н-р, 2, 3, 8, 9), что также может быть связано с тем, что классы перемешаны.

```python
from scipy.cluster import hierarchy

z = hierarchy.linkage(df_scaled, method='ward')
plt.figure(figsize=(20,15))
_ = hierarchy.dendrogram(z)
plt.axhline(y=58, color='r', linestyle='-')
plt.show()
```

По дендрограмме можно выделить одинадцать кластеров, что соответствует реальному числу кластеров.

```python
from sklearn.cluster import AgglomerativeClustering

# кластеризация методов Уорда
n_clusters = len(df.iloc[:, -1].unique()) 
ag = AgglomerativeClustering(n_clusters = n_clusters, linkage = 'ward')
df['ward_label'] = ag.fit_predict(df_scaled)
table = pd.crosstab(df['Class'], df['ward_label'])
table
```

Из кросс-таблицы, полученной по резултатам метода Уорда также видна, что некоторые классы были разбиты на несколько (н-р, 2, 3, 6)

```python
fig, ax = plt.subplots(nrows=1, ncols=3,figsize=(12, 12))

for i, label in enumerate(df.columns[-3:]):
    ax[i].scatter(df_scaled[:, 0], df_scaled[:, 1], c=df[label])
    ax[i].set_title(label)
```

По графикам видно, что кластеры перемешаны.

```python
from sklearn.decomposition import PCA

# применим метод главных компонент
pca = PCA().fit(df_scaled)
n_components = len(pca.singular_values_[pca.singular_values_ > 1])
print(f'Number of components: {n_components} (total number of components: {len(pca.singular_values_)})')
print(f'This number of components explained {sum(pca.explained_variance_ratio_[:n_components]) * 100}% of variance')

pca = PCA(n_components=n_components)
pca.fit(df_scaled)

new_x = pca.transform(df_scaled)
new_x.shape
```

Из 40 признаков получили 37 компонент

```python
# расчитаем инерцию (inertia), искажение (distortion) и метрику силуэта (silhouette)
inertia=[]
distortion = []
silhouette = []
rg = range(2, 14)
for k in rg:
    kmeans = KMeans(n_clusters = k, random_state = 22222).fit(new_x)
    inertia.append(kmeans.inertia_)
    distortion.append(sum(np.min(cdist(new_x,
        kmeans.cluster_centers_,
        'euclidean'),axis = 1))/new_x.shape[0])
    preds = kmeans.predict(new_x)
    silhouette.append(silhouette_score(new_x,
    preds))
print('inertia: ', inertia)
print('distortion: ', distortion)
print('solhouette: ', silhouette)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("Метод локтя для инерции\n",fontsize=16)
plt.scatter(x=[i for i in rg],y=inertia,s=150,edgecolor='k')
plt.plot(rg, inertia);
plt.grid(True)
plt.xlabel("Число кластеров",fontsize=14)
plt.ylabel("Инерция",fontsize=15)
plt.xticks([i for i in rg],fontsize=14)
plt.yticks(fontsize=15)
plt.show()
```

```python
plt.title("Метод локтя для иcкажения\n",fontsize=16)
plt.scatter(x=[i for i in rg],
y=distortion,s=150,edgecolor='k')
plt.plot(rg, distortion);
plt.grid(True)
plt.xlabel("Число кластеров",fontsize=14)
plt.ylabel("Искажение",fontsize=15)
plt.xticks([i for i in rg],fontsize=14)
plt.yticks(fontsize=15)
plt.show()
```

```python
plt.title("Силуэт\n",fontsize=16)
plt.scatter(x=[i for i in rg],
y=silhouette,s=150,edgecolor='k')
plt.plot(rg,silhouette);
plt.grid(True)
plt.xlabel("Число кластеров",fontsize=14)
plt.ylabel("Силуэт",fontsize=15)
plt.xticks([i for i in rg],fontsize=14)
plt.yticks(fontsize=15)
plt.show()
```

В целом метод локтя для инерции и искажения, а также график силуэта дали те же результаты, что и до PCA. Это связано с тем, что оставшиеся компоненты объясняют 100% дисперсии.

```python
n_clusters = len(df.iloc[:, -1].unique())
kmeans = KMeans(n_clusters = 6, random_state = 22222).fit(new_x)
df['kmeans_labels_pca'] = kmeans.labels_
table = pd.crosstab(df['Class'], df['kmeans_labels_pca'])
table
```

Классы все так же остались перемешаны.

```python
from scipy.cluster import hierarchy

z = hierarchy.linkage(new_x, method='ward')
plt.figure(figsize=(20,15))
_ = hierarchy.dendrogram(z)
plt.axhline(y=58, color='r', linestyle='-')
plt.show()
```

По дендрограмме можно выделить 11 кластеров, как и до PCA.

```python
# кластеризация методов Уорда
n_clusters = len(df.iloc[:, -1].unique())
ag = AgglomerativeClustering(n_clusters = n_clusters, linkage = 'ward')
df['ward_label_pca'] = ag.fit_predict(new_x)
table = pd.crosstab(df['Class'], df['ward_label_pca'])
table
```

Кластеры перемешаны.

```python
fig, ax = plt.subplots(nrows=1, ncols=3,figsize=(12, 12))

for i, label in enumerate(["Class", "kmeans_labels_pca", "ward_label_pca"]):
    ax[i].scatter(new_x[:, 0], new_x[:, 1], c=df[label])
    ax[i].set_title(label)
```

После PCA классы остались перемешанными, однако их визуализация по главны мкомпонентам улучшилась и стала более четкой, визуально кластеры стало проще разделить.
