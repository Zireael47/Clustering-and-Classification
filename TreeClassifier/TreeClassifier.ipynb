```python
import pandas as pd
```

```python
df = pd.read_csv("texture.dat")
df.head()
```

```python
#  стандартизация
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(df.iloc[:, :-1])
y = df[' Class']
```

```python
from sklearn.decomposition import PCA

# метод главных компонент
pca = PCA()
pca.fit(X)

# критерий Кайзера
eig_values = pca.singular_values_
print(len(eig_values[eig_values > 1]))
```

```python
pca = PCA(n_components=37)
pca.fit(X)
new_X = pca.transform(X)
new_X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(new_X, y, test_size=0.4, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)

y_pred = tree.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

# точность
accuracy_score(y_test, y_pred)
```

Точность дерева > 91%, что являтся достаточно хорошим результатом
